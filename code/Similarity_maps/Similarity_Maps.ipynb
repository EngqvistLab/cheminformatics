```python
# test av fingerprint sim_maps
from rdkit import Chem
from rdkit.Chem import Draw
from rdkit.Chem.Draw import SimilarityMaps

refmol = Chem.MolFromSmiles('C[C@@H](C(=O)[O-])O')
mol = [Chem.MolFromSmiles('[O-]C(=O)C(O)CCCCCCCCCC'), Chem.MolFromSmiles('CCCCCCCCCCCCCCC(C(=O)[O-])O'), Chem.MolFromSmiles('C(C(=O)[O-])O')]
for Smile in mol:
    fp = SimilarityMaps.GetAPFingerprint(Smile, fpType='normal')
    fp = SimilarityMaps.GetTTFingerprint(Smile, fpType='normal')
    fp = SimilarityMaps.GetMorganFingerprint(Smile, fpType='bv')
    fig, maxweight = SimilarityMaps.GetSimilarityMapForFingerprint(refmol, Smile, SimilarityMaps.GetMorganFingerprint)
    
fig, maxweight
```
