```python
#Test av similaritymaps koppierat från RDkit hemsidan

from rdkit import Chem
mol = Chem.MolFromSmiles('COc1cccc2cc(C(=O)NCCCCN3CCN(c4cccc5nccnc54)CC3)oc21') 
refmol = Chem.MolFromSmiles('CCCN(CCCCN1CCN(c2ccccc2OC)CC1)Cc1ccc2ccccc2c1')

from rdkit.Chem import Draw
from rdkit.Chem.Draw import SimilarityMaps
fp = SimilarityMaps.GetAPFingerprint(mol, fpType='normal')
fp = SimilarityMaps.GetTTFingerprint(mol, fpType='normal')
fp = SimilarityMaps.GetMorganFingerprint(mol, fpType='bv')
fig, maxweight = SimilarityMaps.GetSimilarityMapForFingerprint(refmol, mol, SimilarityMaps.GetMorganFingerprint)
```

```python
#Test av de två första av Martins Smiles med mol2 som referensmolekyl
mol1 = Chem.MolFromSmiles('[O-]C(=O)C(O)CCCCCCCCCC')
mol2 = Chem.MolFromSmiles('C[C@@H](C(=O)[O-])O')


fp = SimilarityMaps.GetAPFingerprint(mol1, fpType='normal')
fp = SimilarityMaps.GetTTFingerprint(mol1, fpType='normal')
fp = SimilarityMaps.GetMorganFingerprint(mol1, fpType='bv')
fig, maxweight = SimilarityMaps.GetSimilarityMapForFingerprint(mol2, mol1, SimilarityMaps.GetMorganFingerprint)
```
